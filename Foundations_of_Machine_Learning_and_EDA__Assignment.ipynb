Question 2:

Explain overfitting and underfitting in ML. How can you detect and prevent
them?

(Hint: Discuss bias-variance tradeoff, cross-validation, and regularization techniques)?

ANSWER:

Overfitting and Underfitting in Machine Learning:


* Overfitting occurs when a machine learning model learns the training data too closely, including noise and unnecessary details.

* As a result, it performs very well on the training data but poorly on new (test) data.

Underfitting:

* Underfitting occurs when a model is too simple to learn the important patterns in the training data.

* Therefore, it performs poorly on both the training and test data.

Detection:

* Compare training and testing performance.

* Overfitting: High training accuracy but low testing accuracy.

* Underfitting: Low training accuracy and low testing accuracy.

* Cross-validation helps detect whether the model generalizes well to unseen data.

Prevention:

* To prevent overfitting:
Apply regularization (L1 or L2) to reduce model complexity.

* Use cross-validation to evaluate model performance.

* Reduce unnecessary model complexity or use more training data.

To prevent underfitting:

* Increase model complexity.

* Train the model for more iterations if required.

* Select features that better represent the data.

Bias–Variance Tradeoff:

* Underfitting: High bias, low variance.

* Overfitting: Low bias, high variance.

* A good model maintains a balance between bias and variance.


Question 3:How would you handle missing values in a dataset?

Explain at least three methods with examples.

(Hint: Consider deletion, mean/median imputation, and predictive modeling)?

ANSWER:

Missing values are empty or unavailable values in a dataset. They should be handled before data analysis to ensure the dataset is complete.

1. Deletion Method:

* In this method, rows or columns containing missing values are removed from the dataset.

Example:

* If a dataset has 100 records and 2 records have missing values, those 2 records can be deleted if they are very few.

2. Mean/Median Imputation:

* Missing numerical values are replaced with the mean or median of the column.

* Mean is used when the data is normally distributed.

* Median is used when the data contains outliers.

Example:

* Age = 20, 22, NaN, 24, 26

* Mean = 23, so the missing value is replaced with 23.

3. Predictive Modeling

* A machine learning algorithm predicts the missing values using other available features in the dataset.

Example:

* If the Salary value is missing, it can be predicted using features such as Age, Education, and Experience.

Question 4:

What is an imbalanced dataset?

Describe two techniques to handle it (theoretical + practical)?

(Hint: Discuss SMOTE, Random Under/Oversampling, and class weights in models)?

ANSWER:

Imbalanced Dataset

An imbalanced dataset is a dataset in which the number of instances in one class is much higher than the number of instances in another class. Because of this imbalance, a machine learning model may become biased toward the majority class.

Example:

* A fraud detection dataset has 9,900 non-fraud transactions and 100 fraud transactions. This is an imbalanced dataset.

Technique 1: SMOTE (Synthetic Minority Over-sampling Technique):

Theory:

* SMOTE increases the number of samples in the minority class by creating new synthetic (artificial) data points based on existing minority class samples.

Practical:

* If a dataset has 900 Class A samples and 100 Class B samples, SMOTE generates new Class B samples so that the classes become more balanced.

Technique 2: Random Oversampling:

Theory:

* Random Oversampling balances the dataset by randomly duplicating existing samples from the minority class until the class distribution becomes balanced.

Practical:

* If a dataset contains 900 majority-class samples and 100 minority-class samples, Random Oversampling duplicates minority-class samples to increase them to 900.

Question 5:

 Why is feature scaling important in ML? Compare Min-Max scaling and
Standardization?

Hint: Explain impact on distance-based algorithms (e.g., KNN, SVM) and gradient
descent.

ANSWER:

Feature Scaling:

* Feature scaling is the process of bringing all features to a similar numerical range so that no feature dominates others because of its larger values. It helps machine learning models learn from all features fairly.

* Feature scaling is especially important for distance-based algorithms like KNN and SVM, where large-value features can dominate distance calculations. It also helps gradient descent converge faster and more efficiently during model training.

| **Min-Max Scaling**                                 | **Standardization**                                                             |
| --------------------------------------------------- | ------------------------------------------------------------------------------- |
| Scales values to a fixed range, usually **0 to 1**. | Transforms data so that it has **mean = 0** and **standard deviation = 1**.     |
| Uses the minimum and maximum values of the feature. | Uses the mean and standard deviation of the feature.                            |
| Best when a fixed range of values is required.      | Best when data has different distributions and is not limited to a fixed range. |




Question 6:

Compare Label Encoding and One-Hot Encoding. When would you prefer
one over the other?

(Hint: Consider categorical variables with ordinal vs. nominal relationships)?

ANSWER:

Label Encoding:

* Label Encoding converts each category into a unique numerical value. Each category is assigned a different integer.

Example:

Low → 0
Medium → 1
High → 2

* It is preferred for ordinal categorical variables, where the categories have a natural order or ranking.

One-Hot Encoding

* One-Hot Encoding converts each category into a separate binary (0 or 1) column. A value of 1 indicates the presence of that category, and 0 indicates its absence.



Question 7:

Google Play Store Dataset:
  
a). Analyze the relationship between app categories and ratings.
Which categories have the highest/lowest average ratings, and what could be the possible reasons?


Dataset: https://github.com/MasteriNeuron/datasets.git

(Include your Python code and output in the code box below.)?

ANSWER:

* The relationship between app categories and ratings can be analyzed by calculating the average rating of each category using the groupby() function.

* Highest Average Rated Categories: Events, Education, and Books & Reference generally have higher average ratings because they provide useful content and have fewer quality issues.
Lowest Average Rated Categories: Dating, Tools, and Communication generally have lower average ratings because users often report advertisements, bugs, and performance issues.




```python
import pandas as pd

# Download the dataset from the provided GitHub link
!wget https://raw.githubusercontent.com/MasteriNeuron/datasets/main/googleplaystore.csv

# Load the dataset
df = pd.read_csv("googleplaystore.csv")

# Keep only valid ratings
df["Rating"] = pd.to_numeric(df["Rating"], errors="coerce")
df = df.dropna(subset=["Rating"])

# Calculate average rating for each category
avg_rating = df.groupby("Category")["Rating"].mean().sort_values(ascending=False)

# Display average ratings
print(avg_rating)

# Highest and lowest average rated categories
print("\nHighest Average Rated Category:")
print(avg_rating.head(1))

print("\nLowest Average Rated Category:")
print(avg_rating.tail(1))
```

Question 8: Titanic Dataset

a) Compare the survival rates based on passenger class (Pclass). Which class had the highest
survival rate, and why do you think that happened?

b) Analyze how age (Age) affected survival. Group passengers into children (Age < 18) and
adults (Age ≥ 18). Did children have a better chance of survival?

Dataset: https://github.com/MasteriNeuron/datasets.git

(Include your Python code and output in the code box below.)?

ANSWER:

a) Survival Rate Based on Passenger Class (Pclass)

The survival rate was different for each passenger class.

* 1st Class (Pclass = 1) had the highest survival rate.

* 2nd Class had a moderate survival rate.

* 3rd Class had the lowest survival rate.

Reason:

* Passengers in 1st Class were generally given priority during evacuation and had better access to lifeboats.

b) Effect of Age on Survival

Passengers are grouped as:

* Children: Age < 18
* Adults: Age ≥ 18

After comparing survival rates:

* Children had a better chance of survival than adults.

Reason:

* During the rescue, children were given higher priority for lifeboats.


```python
import pandas as pd

# Load dataset
df = pd.read_csv("https://raw.githubusercontent.com/MasteriNeuron/datasets/main/titanic.csv")

# -------------------------------
# a) Survival rate by Passenger Class
# -------------------------------
pclass_survival = df.groupby("Pclass")["Survived"].mean() * 100

print("Survival Rate by Passenger Class (%)")
print(pclass_survival.round(2))

# -------------------------------
# b) Survival rate by Age Group
# -------------------------------
df_age = df.dropna(subset=["Age"]).copy()

df_age["Age_Group"] = df_age["Age"].apply(
    lambda x: "Children" if x < 18 else "Adults"
)

age_survival = df_age.groupby("Age_Group")["Survived"].mean() * 100

print("\nSurvival Rate by Age Group (%)")
print(age_survival.round(2))
```

Question 9: Flight Price Prediction Dataset
a) How do flight prices vary with the days left until departure? Identify any exponential price
surges and recommend the best booking window.

b)Compare prices across airlines for the same route (e.g., Delhi-Mumbai). Which airlines are
consistently cheaper/premium, and why?

Dataset: https://github.com/MasteriNeuron/datasets.git

(Include your Python code and output in the code box below.)?


ANSWER:

a) Flight Price vs. Days Left Until Departure:

Flight prices generally increase as the number of days left until departure decreases.

* When there are more days left, ticket prices are usually lower.

* A sharp (exponential) price increase is commonly observed during the last 7–10 days before departure.

Recommendation:

* The best booking window is about 20–40 days before departure, when prices are generally lower.

b) Price Comparison Across Airlines (Delhi → Mumbai):

For the Delhi → Mumbai route:

* Low-cost airlines such as Indigo and AirAsia usually offer lower ticket prices.

* Premium airlines such as Vistara and Air India generally have higher ticket prices.

Reason:

* Premium airlines usually provide additional services and better travel facilities, resulting in higher fares.

```python
import pandas as pd

# Load dataset
df = pd.read_csv(
    "https://raw.githubusercontent.com/MasteriNeuron/datasets/main/flight_price.csv"
)

# -------------------------------
# a) Average Price by Days Left
# -------------------------------
days_price = df.groupby("days_left")["price"].mean().sort_index()

print("Average Price by Days Left")
print(days_price.round(2))

# -------------------------------
# b) Airline Price Comparison
#    (Delhi -> Mumbai)
# -------------------------------
route = df[
    (df["source_city"] == "Delhi") &
    (df["destination_city"] == "Mumbai")
]

airline_price = route.groupby("airline")["price"].mean().sort_values()

print("\nAverage Price (Delhi -> Mumbai)")
print(airline_price.round(2))
```

Question 10: HR Analytics Dataset

a). What factors most strongly correlate with employee attrition?

Use visualizations to show key
drivers. (e.g., satisfaction, overtime, salary).

b). Are employees with more projects more likely to leave?

Dataset: hr_analytics


ANSWER


a) Factors Correlated with Employee Attrition

The factors that show the strongest relationship with employee attrition are:

* Low job satisfaction → Higher attrition.

* Overtime → Employees working overtime leave more frequently.

* Low salary → Employees with lower salaries have higher attrition.

These factors can be visualized using count plots and box plots.

b) Relationship Between Number of Projects and Attrition

Employees handling more projects are generally more likely to leave the company.

This indicates that a higher workload is associated with increased employee attrition.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Download dataset
!wget https://raw.githubusercontent.com/MasteriNeuron/datasets/main/hr_analytics.csv

# Load dataset
df = pd.read_csv("hr_analytics.csv")

# ------------------------------------
# a) Visualizations for Attrition
# ------------------------------------

# Job Satisfaction vs Attrition
sns.countplot(x="satisfaction_level", hue="left", data=df)
plt.title("Satisfaction Level vs Attrition")
plt.show()

# Overtime (average monthly hours) vs Attrition
sns.boxplot(x="left", y="average_montly_hours", data=df)
plt.title("Average Monthly Hours vs Attrition")
plt.show()

# Salary vs Attrition
sns.countplot(x="salary", hue="left", data=df)
plt.title("Salary vs Attrition")
plt.show()

# ------------------------------------
# b) Number of Projects vs Attrition
# ------------------------------------

project_attrition = df.groupby("number_project")["left"].mean() * 100

print("Attrition Rate by Number of Projects (%)")
print(project_attrition.round(2))
```
